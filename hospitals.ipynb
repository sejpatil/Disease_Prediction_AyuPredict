```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
```

```python
data = pd.read_csv(r"C:\Users\Sejal\OneDrive\Desktop\Hospitals_list.csv")

df = pd.DataFrame(data)
```

```python
df['Features'] = df['City'] + ' ' + df['Hospital'] + ' ' + df['State']
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['Features'])
```

```python
knn = NearestNeighbors(n_neighbors=50, algorithm='brute', metric='cosine')
knn.fit(X)
```

```python
def get_hospitals(user_location, k=3):
    # Transform user input
    user_input = vectorizer.transform([user_location])

    # Find nearest neighbors
    _, indices = knn.kneighbors(user_input)

    # Get hospital information
    hospitals = df.iloc[indices[0]]['Hospital'].values
    addresses = df.iloc[indices[0]]['LocalAddress'].values

    result = []
    for i in range(len(hospitals)):
        result.append((hospitals[i], addresses[i]))

    return result
```

```python
user_location = 'Pune'
result = get_hospitals(user_location)
print("Hospitals in", user_location)
for hospital, address in result:
    print(hospital, "-", address)
```
